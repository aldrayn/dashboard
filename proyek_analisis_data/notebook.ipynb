**Menampilkan banyaknya nilai outlier dari setiap kolom**  

Pada proses ini, saya ingin mengidentifikasi nilai outlier dalam setiap kolom. Untuk melakukan ini, saya membuat fungsi bernama count_outlier yang menggunakan kriteria Interquartile Range (IQR).

Fungsi ini bekerja dengan cara berikut :

- Menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3) untuk kolom yang diberikan.
- Menghitung IQR dengan rumus: IQR = Q3 - Q1.
- Menentukan batas bawah dan batas atas untuk nilai normal: batas bawah = Q1 - 1.5 * IQR dan batas atas = Q3 + 1.5 * IQR.
- Menghitung jumlah nilai yang berada di luar batas ini.  
  
Dengan menggunakan fungsi ini, saya dapat menampilkan banyaknya nilai outlier untuk setiap kolom dalam dataset. Ini penting untuk memastikan bahwa analisis saya tidak dipengaruhi oleh nilai-nilai ekstrem yang mungkin merusak hasil analisis.

```python
def count_outlier(data) :
    
    outlier_df = pd.DataFrame(columns = ['num_of_outliers'])
    
    for num_col in data.select_dtypes('number').columns :
    
        q1  = np.percentile(data[num_col], 25)
        q3  = np.percentile(data[num_col], 75)
        iqr = q3 - q1

        upper_limit = q3 + 1.5 * iqr
        lower_limit = q1 - 1.5 * iqr
    
        outlier =  data[(data[num_col] > upper_limit) | (data[num_col] < lower_limit)].shape[0]
        
        outlier_df.loc[num_col, 'num_of_outliers'] = outlier
    
    return outlier_df

count_outlier(day_df)
```

Menetapkan inplace=True memastikan bahwa perubahan diterapkan langsung ke DataFrame yang ada.Menghapus data duplikat adalah langkah krusial untuk memastikan bahwa analisis saya tidak terpengaruh oleh data yang berulang, sehingga hasil analisis menjadi lebih akurat dan dapat diandalkan

Dengan perintah `.dropna(inplace=True)`, saya dapat menghapus semua baris yang mengandung nilai null. Menghapus missing value adalah langkah penting untuk menjaga kualitas data, sehingga analisis yang saya lakukan dapat menghasilkan hasil yang lebih akurat dan representatif.

**Handling Outlier**  

- Pada proses ini, saya melakukan penanganan terhadap nilai outlier dalam dataset day_df dengan mengganti nilai-nilai ekstrem menggunakan batas atas dan batas bawah. Metode yang digunakan adalah menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3) untuk menentukan Interquartile Range (IQR), kemudian mengidentifikasi outlier berdasarkan batas-batas ini.

- Setelah mendapatkan batas bawah dan batas atas, saya mengganti nilai yang melebihi batas tersebut dengan nilai upper bound untuk outlier positif dan nilai lower bound untuk outlier negatif. Tujuan dari proses ini adalah untuk memastikan bahwa analisis saya tidak terpengaruh oleh nilai ekstrem yang dapat merusak hasil, sehingga data yang saya gunakan menjadi lebih bersih dan representatif.

```python
kolom_outlier = ['windspeed', 'hum']

def handling_outlier(data) :
    q1 = (data).quantile(0.25)
    q3 = (data).quantile(0.75)
    
    iqr = q3 - q1
    
    upper_bond = q3 + (1.5 * iqr)
    lower_bond = q1 - (1.5 * iqr)
    
    lower_than = data < lower_bond
    more_than  = data > upper_bond
    
    data[more_than]  = upper_bond
    data[lower_than] = lower_bond

    return data
```

**Perbandingan jumlah outlier pada sebelum dan sesudah dilakukan penanganan**

```python
before = count_outlier(day_df)
before.rename(columns = {'num_of_outliers' : 'Before'}, inplace = True) 

for col in kolom_outlier :
    day_df[col] = handling_outlier(day_df[col])
    
after = count_outlier(day_df) # Sesudah menangani outlier
after.rename(columns = {'num_of_outliers' : 'After'}, inplace = True) 

mix = before.join(after, how = 'inner')
mix
```

Untuk keperluan dashboard, Data yang akan dipakai adalah data yang skalanya sudah dikembalikan menjadi skala data asli (tanpa encoding atau normalisasi). Tujuannya adalah agar pembaca dapat memahami grafik dengan mudah kerana sudah mengetahui nilai asli dari masing-masing variable

```python
def convert_to_original_scale(data) :
    data['season'] = data['season'].map({   1 : 'springer', 
                                                2 : 'summer', 
                                                3 : 'fall', 
                                                4 : 'winter'
    })

    data['yr']   = np.where(data['yr'] == 0, '2011', '2012')

    data['mnth'] = data['mnth'].map({   1 : 'January',    2 : 'February', 
                                            3 : 'March',      4 : 'April',
                                            5 : 'May',        6 : 'June', 
                                            7 : 'July',       8 : 'August', 
                                            9 : 'September', 10 : 'October', 
                                            11: 'November',  12 : 'December'})

    data['holiday'] = np.where(data['holiday'] == 0, 'No', 'Yes')
    data['weekday'] = data['weekday'].map({ 
                                                1 : 'Monday', 
                                                2 : 'Tuesday', 
                                                3 : 'Wednesday', 
                                                4 : 'Thrusday', 
                                                5 : 'Friday',
                                                6 : 'Saturday',
                                                0 : 'Sunday'
    })

    data['workingday'] = data['workingday'].map({0 : 'No', 1 : 'Yes'})
    data['weathersit'] = data['weathersit'].map({1 : 'Normal',   2 : 'Moderate', 3 : 'Caution', 4 : 'Extreme'})
    
    data['temp']  = round(41  * data['temp'])
    data['atemp'] = round(50  * data['atemp'])
    data['hum']   = round(100 * data['hum'])

    data['windspeed'] = round(67  * data['windspeed'])
    
    return data


day_df['date']  = [d[-2:] for d in day_df['dteday']]
ori_day_df = day_df.copy()
ori_day_df  = convert_to_original_scale(ori_day_df)

hour_df['date']  = [d[-2:] for d in hour_df['dteday']]
ori_hour_df = hour_df.copy()
ori_hour_df = convert_to_original_scale(ori_hour_df)
```

**Numerik - Numerik**  

Untuk kolom numerik yang saling berhubungan, saya menggunakan koefisien korelasi Pearson. Metode ini memungkinkan saya untuk mengukur kekuatan dan arah hubungan linier antara dua variabel numerik.

```python
num_num = ori_day_df.select_dtypes('number').corr()
num_num
```

**Numerik - Biner**  

Untuk menguji hubungan antara kolom numerik dan biner, saya menerapkan koefisien korelasi Point Biserial. Library yang digunakan adalah `scipy`.

```python
from scipy.stats import pointbiserialr

num_bin = pd.DataFrame()

numeric_col = ori_day_df.select_dtypes('number').columns
biner_col   = ['yr', 'holiday', 'workingday']

for num in numeric_col :
    for bin in biner_col :
        r_pb, p_value = pointbiserialr(day_df[bin], day_df[num])
        
        num_bin.loc[num, bin] = r_pb
        
num_bin
```

**Biner - Biner**  

Untuk hubungan antara dua kolom biner, saya menggunakan koefisien Phi, yang cocok untuk data biner.

```python
from scipy.stats import chi2_contingency

bin_bin = pd.DataFrame()

biner_col   = ['yr', 'holiday', 'workingday', 'weathersit']

for a in biner_col :
    for b in biner_col :
        contingency_table = pd.crosstab(day_df[a], day_df[b])

        chi2, p, dof, expected = chi2_contingency(contingency_table)

        phi_coefficient = np.sqrt(chi2 / day_df.shape[0])
        
        bin_bin.loc[a, b] = phi_coefficient

bin_bin
```

**Multi kategori - Biner**  

Ketika menganalisis hubungan antara kolom biner dan kolom multi kategorikal, saya menggunakan Cramér's V.

```python
multi_biner = pd.DataFrame()

multi_cat_col = [col for col in ori_day_df.drop(columns = 'dteday').select_dtypes('object') if ori_day_df[col].nunique() > 2]

biner_col = ['yr', 'holiday', 'workingday', 'weathersit']

for a in multi_cat_col :
    
    for b in biner_col :    
        contingen_table = pd.crosstab(index = day_df[a], columns = day_df[b])
        
        chi2, p, dof, expected_freq = chi2_contingency(contingen_table)
        
        n = np.sum(np.sum(contingen_table))
        
        v = np.sqrt(chi2/(n * (min(contingen_table.shape) - 1)))
    
        multi_biner.loc[a, b] = v

multi_biner
```

**Multi kategori - Multi Kategori**  

Untuk mengeksplorasi hubungan antara dua kolom multi kategorikal, saya juga menerapkan Cramér's V.

```python
multi_multi = pd.DataFrame()

biner_col = ['yr', 'holiday', 'workingday', 'weathersit']

for a in multi_cat_col :
    
    for b in multi_cat_col :    
        contingen_table = pd.crosstab(index = day_df[a], columns = day_df[b])
        
        chi2, p, dof, expected_freq = chi2_contingency(contingen_table)
        
        n = np.sum(np.sum(contingen_table))
        
        v = np.sqrt(chi2/(n * (min(contingen_table.shape) - 1)))
    
        multi_multi.loc[a, b] = v

multi_multi
```

```python
# Daftar kolom yang bertipe data continuous
continuous_columns = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']

fig, ax = plt.subplots(2, 4, figsize = (10, 8.4))

baris, kolom = 0, 0

# Membuat dataframe kosong untuk tempat menyimpan nilai skew
shape_data = pd.DataFrame() 

# Menampilkan histogram dari semua kolom continuous
for idx, cont in enumerate(continuous_columns) :
    
    if idx % 2 == 0 :
        color = warna_default = plt.rcParams['axes.prop_cycle'].by_key()['color'][0]
    else :
        color = 'orange'
        
    # Memasukkan nilai skew dan kurt ke dataframe yang akan ditampilkan di halaman kosong
    skewn = ori_hour_df[cont].skew()
    kurto = ori_hour_df[cont].kurt()
    
    shape_data.loc[cont, 'skew'] = skewn
    shape_data.loc[cont, 'kurt'] = kurto
    
    ax[baris, kolom].hist(ori_hour_df[cont], edgecolor = 'black', color = color)
    ax[baris, kolom].set_title(cont)
    
    if kolom < 3 :
        kolom += 1
    else :
        baris +=1
        kolom = 0
    
ax[1, 3].axis('off') # Menghilangkan sumbu

# Membuat tabel yang berisi nilai-nilai di shape_data
table = ax[1, 3].table(cellText = [[col, round(shape_score[0], 2), round(shape_score[1], 2)] for col, shape_score in zip(shape_data.index, shape_data.values)], colLabels = ['', 'skew', 'kurt'], cellLoc = 'center', loc = 'center')

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.4, 2.2) 

plt.tight_layout() ;
```

#### **Hubungan Antar Kolom**

**Numerik - Numerik**  

Untuk kolom numerik yang saling berhubungan, saya menggunakan koefisien korelasi Pearson. Metode ini memungkinkan saya untuk mengukur kekuatan dan arah hubungan linier antara dua variabel numerik.

```python
num_num = ori_hour_df.select_dtypes('number').corr()
num_num
```

**Numerik - Biner**  

Untuk menguji hubungan antara kolom numerik dan biner, saya menerapkan koefisien korelasi Point Biserial. Library yang digunakan adalah `scipy`.

```python
from scipy.stats import pointbiserialr

num_bin = pd.DataFrame()

numeric_col = ori_hour_df.select_dtypes('number').columns
biner_col   = ['yr', 'holiday', 'workingday']

for num in numeric_col :
    for bin in biner_col :
        r_pb, p_value = pointbiserialr(hour_df[bin], hour_df[num])
        
        num_bin.loc[num, bin] = r_pb
        
num_bin
```

**Biner - Biner**  

Untuk hubungan antara dua kolom biner, saya menggunakan koefisien Phi, yang cocok untuk data biner.

```python
from scipy.stats import chi2_contingency

bin_bin = pd.DataFrame()

biner_col   = ['yr', 'holiday', 'workingday']

for a in biner_col :
    for b in biner_col :
        contingency_table = pd.crosstab(hour_df[a], hour_df[b])

        chi2, p, dof, expected = chi2_contingency(contingency_table)

        phi_coefficient = np.sqrt(chi2 / hour_df.shape[0])
        
        bin_bin.loc[a, b] = phi_coefficient

bin_bin
```

**Multi kategori - Biner**  

Ketika menganalisis hubungan antara kolom biner dan kolom multi kategorikal, saya menggunakan Cramér's V.

```python
multi_biner = pd.DataFrame()

multi_cat_col = [col for col in ori_hour_df.drop(columns = 'dteday').select_dtypes('object') if ori_hour_df[col].nunique() > 2]

biner_col = ['yr', 'holiday', 'workingday']

for a in multi_cat_col :
    
    for b in biner_col :    
        contingen_table = pd.crosstab(index = hour_df[a], columns = hour_df[b])
        
        chi2, p, dof, expected_freq = chi2_contingency(contingen_table)
        
        n = np.sum(np.sum(contingen_table))
        
        v = np.sqrt(chi2/(n * (min(contingen_table.shape) - 1)))
    
        multi_biner.loc[a, b] = v

multi_biner
```

**Multi kategori - Multi Kategori**  

Untuk mengeksplorasi hubungan antara dua kolom multi kategorikal, saya juga menerapkan Cramér's V.

```python
multi_multi = pd.DataFrame()

for a in multi_cat_col :
    
    for b in multi_cat_col :    
        contingen_table = pd.crosstab(index = hour_df[a], columns = hour_df[b])
        
        chi2, p, dof, expected_freq = chi2_contingency(contingen_table)
        
        n = np.sum(np.sum(contingen_table))
        
        v = np.sqrt(chi2/(n * (min(contingen_table.shape) - 1)))
    
        multi_multi.loc[a, b] = v

multi_multi
```

### **hour_df**

#### **Pertanyaan 3 : Bagaimana tren penyewaan sepeda setiap musim berdasarkan jamnya ?**

```python
fig, ax = plt.subplots(2, 2, figsize = (10, 5)) 

fig.suptitle('Rata-rata jumlah pelanggan tiap jam di setiap musim')

baris, kolom = 0, 0

for s in np.unique(ori_hour_df['season']) :
    
    count = ori_hour_df[ori_hour_df['season'] == s].groupby('hr')['cnt'].mean()
    
    # Mengambil index dengan nilai 3 terbesar
    top_three = count.nlargest(3)
    
    warna_default = plt.rcParams['axes.prop_cycle'].by_key()['color'][0] # --> Digunakan untuk mengambil warna default pertama dari siklus warna matplotlib
    
    # warna top three --> orange, selain itu : warna default matplotlib (biru)
    colors = ['orange' if idx in top_three else warna_default for idx in count.index]
    
    ax[baris, kolom].bar(range(0, 24), count, color = colors)
    ax[baris, kolom].set_title(f'Musim {s}')
    
    if baris == 1 :
        ax[baris, kolom].set_xlabel('Jam')
    
    if kolom == 0 :
        ax[baris, kolom].set_ylabel('Count')
    
    ax[baris, kolom].set_ylim(0, 600)
    
    if kolom < 1 :
        kolom += 1
    else :
        baris += 1
        kolom = 0
    
plt.tight_layout()
plt.show()
```
